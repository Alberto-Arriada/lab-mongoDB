### 2. All the companies that have more than 5000 employees. Limit the search to 20 companies and sort them by **number of employees**.

```python
for x in db.COMPANIES.find(filter={'number_of_employees':{'$gt':5000}},
                           sort=[('number_of_employees',1)], limit=20):
    print(x,'\n')

#print(list(db['COMPANIES'].find(filter={'number_of_employees':{'$gt':5000}},sort=[('number_of_employees',1)], limit=20)))
```

### 3. All the companies founded between 2000 and 2005, both years included. Retrieve only the `name` and `founded_year` fileds.

```python
for x in db.COMPANIES.find({'$and':[{'founded_year':{'$gte':2000}},
                                    {'founded_year':{'$lte':2005}}]}, 
                           projection={'founded_year':1,'name':1,'_id':0}):
     print(x,'\n')
#print(list(db['companies'].find({'$and':[{'founded_year':{'$gte':2000}},{'founded_year':{'$lte':2005}}]},projection={'founder_year':1,'_id':0})))
```

### 12. All the companies that have been 'deadpooled' after the third year.

```python
for i,x in enumerate(db.COMPANIES.find(filter={'$and':[{'$expr':{'$gt':['$deadpoll_year',
                                                                       {'$add':['$founded_year',3]}]}},
                                                       {'founded_year':{'$gte':0}}]},limit=(50),
                                      projection={'name':1,'_id':0,'deadpool_year':1,'founded year':1})):
    print(x,'\n')
```
